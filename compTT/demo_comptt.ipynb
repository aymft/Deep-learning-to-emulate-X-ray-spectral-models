```python
from comptt import comptt_spectra
import matplotlib.pyplot as plt
import numpy as np
import scienceplots
from matplotlib import rcParams

# Input parameters
T0, kT, tau = 0.5, 2.0, 1.2
spectrum = comptt_spectra(T0, kT, tau)

# Energy grid (assumed known from training data)
energy = np.geomspace(0.1, 50, len(spectrum))

plt.style.use(['science', 'no-latex'])
rcParams['figure.dpi'] = 1000
rcParams['font.size'] = 9    
rcParams['axes.titlesize'] = 9
rcParams['axes.labelsize'] = 9
rcParams['xtick.labelsize'] = 7
rcParams['ytick.labelsize'] = 7
rcParams['legend.fontsize'] = 9
rcParams['lines.linewidth'] = 1.0
rcParams['grid.linewidth'] = 0.5
plt.semilogx(energy, spectrum)
plt.xlabel("Energy (keV)")
plt.ylabel("Flux")
plt.title(rf"$T_0={T0}, kT={kT}, \tau={tau}$")
plt.tight_layout()
plt.show()
```
