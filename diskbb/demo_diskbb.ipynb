```python
from diskbb import diskbb_spectra
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import scienceplots

temperature = 7.5
spectrum = diskbb_spectra(temperature)
energy = np.geomspace(0.01, 50, len(spectrum))

plt.style.use(['science', 'no-latex'])

plt.semilogx(energy, spectrum, label = f"T = {temperature} keV" )
plt.xlabel("Energy (keV)")
plt.ylabel("Flux")
plt.legend()
plt.show()
```
